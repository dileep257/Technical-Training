```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
flights = sb.load_dataset("flights")
flights.head()
```

```python
may_flights = flights.query("month == 'May'")
sb.scatterplot(data=may_flights, x="year", y="passengers")
```

```python
sb.lineplot(data=may_flights, x="year", y="passengers")
```

```python
sb.set_style('whitegrid')
sb.lmplot(x ='year', y ='passengers', data = flights)
```
